## Bar Charts for Categorical Variables
+ Sex
+ Pclass
+ Sibsp
+ Parch
+ Cabin
+ Embarked

```python
def bar_charts(feature):
    survival = train_data[train_data['Survived'] == 1][feature].value_counts() # get the number of survivals
    death = train_data[train_data['Survived'] == 0][feature].value_counts() # get the number of deaths
    print('Survival:', survival)
    print('Death:', death)

    df = pd.DataFrame([survival, death])
    print(df)

    df.index = ['Survived', 'Death']

    fig = plt.figure(figsize=(10, 5))

    df.plot(kind = 'bar', stacked = True, ax = fig.gca())
    plt.show()
```

```python
bar_charts('Sex')
```

The bar charts seem to infer that women is more likely to survive than men.

It turns out that 1st class is most likely to survive, while 3rd class is most likely to die.

From the bar chart, it tunrs out that the number of people coming to Titanic without a sibling led in both survival and death numbers.

```python
bar_charts('Parch')
```

From the bar chart, it tunrs out that the number of people coming to Titanic without a parent/children led in both survival and death numbers.

```python
def plot_facet(feature, xmin = None, xmax = None):
    facet = sns.FacetGrid(train_data, hue="Survived", aspect=4)
    facet.map(sns.kdeplot,feature,shade= True)
    facet.set(xlim=(0, train_data[feature].max()))
    facet.add_legend() 
    if (xmin != None and xmax != None):
        plt.xlim(xmin, xmax)
    plt.show()
```

## Binning
Binning/Converting numerical age to categorical variable  
Feature vector map:  
child: 0  
young: 1  
adult: 2  
mid-age: 3  
senior: 4  

```python
for dataset in train_test_data:
    dataset.loc[dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 26), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 26) & (dataset['Age'] <= 36), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 36) & (dataset['Age'] <= 62), 'Age'] = 3
    dataset.loc[dataset['Age'] > 62, 'Age'] = 4
```

# Cross Validation (k-fold)

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
k_fold
```

```python
f = open('score_knn.txt', 'w')
for i in range(20):
    f.write('Number of neighbors: ' + str(i+1) + ', ')
    clf = KNeighborsClassifier(n_neighbors=i + 1)
    scoring = 'accuracy'
    score = cross_val_score(clf, train_data_drop, target_data, cv=k_fold, n_jobs=1, scoring=scoring)
    f.write('Mean: ' + str(round(np.mean(score)*100, 2)) + '%')
    f.write('\n')
f.close()
```

```python
clf_list = [KNeighborsClassifier(n_neighbors = 13),
            DecisionTreeClassifier(),
            RandomForestClassifier(n_estimators=13),
            GaussianNB(),
            SVC(),
            ExtraTreeClassifier(),
            GradientBoostingClassifier(n_estimators=10, learning_rate = 1,max_features=3, max_depth =3, random_state = 10),
            AdaBoostClassifier(),
            ExtraTreesClassifier()]
```

```python
def model_fit(model_list):
    scores = []
    scoring = 'accuracy'
    for i in range(len(clf_list)):
        score =cross_val_score(clf_list[i], train_data_drop, target_data, cv=k_fold, n_jobs=1, scoring=scoring)
        print('Score of model ', i, round(np.mean(score)*100, 2))
        scores.append(round(np.mean(score)*100, 2))
    
    return scores
```

```python
scores = model_fit(clf_list)
```

```python
imax = np.argmax(scores)
imax
```

```python
selected_model = clf_list[imax]
selected_model.fit(train_data_drop, target_data)
```

```python
test = test_data.drop("PassengerId", axis=1).copy()
```

```python
prediction = selected_model.predict(test)
```

```python
test['Survived'] = prediction
test['PassengerId'] = test_data['PassengerId']
test
```
